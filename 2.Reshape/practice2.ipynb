## 8

```python
OC = ['Australasia']
NA = ['Canada', 'United States', 'Mexico']
SA = ['Cuba', 'Jamaica']
EU = ['Czech', 'Yugoslavia', 'Germany', 'United Team of Germany', 'East Germany', 'West Germany', 'Great Britain', 'Russia', 'Russian Empire', 'Soviet Union']
```

```python
medals.loc[medals.index.isin(OC), 'continent'] = 'OC'
medals.loc[medals.index.isin(NA), 'continent'] = 'NA'
medals.loc[medals.index.isin(SA), 'continent'] = 'SA'
medals.loc[medals.index.isin(EU), 'continent'] = 'EU'
medals.loc[medals['continent'].isna(), 'continent'] = 'OT'
```

```python
medals.isna().sum()
```

## 9

```python
medals = medals.set_index(['continent', 'Country'])
medals.head()
```

## 10

```python
medals.sort_values(by=['continent', 'Country'], ascending=[False, True])
```

## 11

```python
medals.columns = pd.MultiIndex.from_tuples(
    [
        ('Code', ''),
        ('Summer', 'Game'), 
        ('Summer', 'Gold'), 
        ('Summer', 'Silver'), 
        ('Summer', 'Bronze'),
        ('Winter', 'Game'), 
        ('Winter', 'Gold'), 
        ('Winter', 'Silver'), 
        ('Winter', 'Bronze'),
    ],
)
medals = medals.drop('Code', axis='columns')
medals
```

## 12

```python
medals = medals.stack(level=0)
medals
```

## 13

```python
medals.reset_index()
medals.columns
```

```python
medals = medals.reset_index()
long = medals.melt(
    id_vars=["continent","Country", "level_2", "Game"],
    var_name="Medal", 
    value_name="Count"
).rename(columns={"level_2": "Season"})
long
```

## 14

```python
long.query('Medal == "Gold" and Season == "Summer"').groupby('continent')['Count'].sum().sort_values(ascending=False)
```
