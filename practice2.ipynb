```python
# medals = medals.join(drinks['continent'])
medals = medals.merge(drinks[['continent']], left_index=True, right_index=True)
medals
```

## 7

```python
medals[medals['continent'].isna()]
old_medals = medals
```

## 8

```python
OC = ['Australasia']
NA = ['Canada', 'United States', 'Mexico']
SA = ['Cuba', 'Jamaica']
EU = ['Czech', 'Yugoslavia', 'Germany', 'Great Britain', 'Russia', 'Soviet']
```

```python
medals.loc[medals.index.isin(OC), 'continent'] = 'OC'
medals.loc[medals.index.isin(NA), 'continent'] = 'NA'
medals.loc[medals.index.isin(SA), 'continent'] = 'SA'
medals.loc[medals.index.isin(EU), 'continent'] = 'EU'
# fill the rest NaN with 'OT'
medals['continent'].fillna('OT', inplace=True)
```

```python
medals.isna().sum()
```

## 9

```python
medals = medals.set_index(['continent', 'Country'])
medals.head()
```

## 10

```python
medals.sort_values(by=['continent', 'Country'])
```

## 11

```python
medals.columns = pd.MultiIndex.from_tuples(
    [
        ('Code', 'Code'),
        ('Summer', 'Game'), 
        ('Summer', 'Gold'), 
        ('Summer', 'Silver'), 
        ('Summer', 'Bronze'),
        ('Winter', 'Game'), 
        ('Winter', 'Gold'), 
        ('Winter', 'Silver'), 
        ('Winter', 'Bronze'),
    ],
    names=['Season', 'Medal']
)
medals
```
