# 19. จำนวนเหรียญทั้งหมดที่แจกไปในโอลิมปิกฤดูหนาวรวมแลวมีกเหรียญ

# 20. หาค่าเฉลี่ยของจำนวนเหรียญทองฤดูหนาวและฤดูร้อนต่อประเทศ

```python
columns = medals[['SummerGold', 'WinterGold']]
result = pd.DataFrame()
result['Country'] = medals['Country']
result['Mean Summer Gold'] = medals[['SummerGold']].mean(axis=1)
result['Mean Winter Gold'] = medals[['WinterGold']].mean(axis=1)
result.set_index('Country').sort_values(by=['Mean Summer Gold', 'Mean Winter Gold'], ascending=[False, False])
```

# 21. แสดงจำนวนเหรียญทั้งหมด (รวมฤดูร้อนและฤดูหนาว) ที่แต่ละประเทศได้ 

```python
total_medal = medals.iloc[:, [3, 4, 5, 7, 8, 9]].sum(axis=1).rename('Total Medal')
pd.concat([medals['Country'], total_medal], axis=1).set_index('Country').sort_values(by='Total Medal', ascending=False)
```

# 22. แสดงประเทศที่ได้เหรียญรวมทุกประเภทของโอลิมปิกฤดูหนาวมากกว่าเหรียญรวมโอลิมปิกฤดูร้อน 
ให้คอลัมน์ของตารางนี้แสดงชื่อประเทศ เหรียญรวมฤดร้อน เหรียญรวมฤดหนาว

```python
result = pd.DataFrame(medals['Country'])
result['Summer'] = medals[['SummerGold', 'SummerSilver', 'SummerBronze']].sum(axis=1)
result['Winter'] = medals[['WinterGold', 'WinterSilver', 'WinterBronze']].sum(axis=1)
result.query('Winter > Summer').set_index('Country')
```

# 23. แสดงประเทศที่ไม่เคยได้เหรียญใดของโอลิมปิกฤดูร้อนแต่ไดเหรียญใดของฤดหนาว หรือ ได้แต่เหรียญฤดู
ร้อนไมเคยได้ฤดูหนาว

```python
result = pd.DataFrame(medals['Country'])
result['Summer'] = medals[['SummerGold', 'SummerSilver', 'SummerBronze']].sum(axis=1)
result['Winter'] = medals[['WinterGold', 'WinterSilver', 'WinterBronze']].sum(axis=1)
result.query('Summer > 0 and Winter == 0 or Winter > 0 and Summer == 0').set_index('Country').sort_values(by='Summer', ascending=False)
```

# 26. ให้สร้างคอลัมน์ใหม่ในตาราง medals ให้ชื่อวา score โดยคำนวณคะแนนรวมจากเหรียญที่แต่ละประเทศ
ได้ (ทั้งฤดูร้อนและหนาว) โดยให้เหรียญทองได้ 3 คะแนน เหรียญเงินได 2 คะแนน เหรียญทองแดงได้ 1
คะแนน แล้วแสดง้อมล 10 ประเทศแรกที่ไดคะแนนเยอะที่สด

```python
medals['Score'] = medals.iloc[:, [3, 4, 5, 7, 8, 9]].multiply([3, 2, 1, 3, 2, 1]).sum(axis=1)
medals.sort_values(by='Score', ascending=False)[['Country', 'Score']].set_index('Country').head(10)
```

# 27. การเปรียบเทียบคะแนนจากเหรียญทั้งหมดอาจจะไมแฟร์ เพราะแต่ละประเทศเข้ารวมการแข่งขันไม่เท่ากัน ดังนั้นให้สร้างคอลัมน์ใหม่ในตาราง medals ชื่อวา perGame โดยคำนวคะแนนต่อการเข้าร่วมแข่งขัน (คะแนนคนวณเหมือนเดม แต่หารด้วยจนวนเกมที่เข้าแข่งขัน แยกกันระหว่างฤดูร้อนกับฤดูหนาว ค่าperGame จะเปนผลบวกของคะแนนของฤดรอนกบฤดหนาว) แสดงขอมล 10 ประเทศที่ได้คะแนนน้อยที่สุด
1. ระวัง ตรวจสอบผลลพธ์ให้ดี มี surprise นะข้อนี้

```python
winter_score = medals[['WinterGold', 'WinterSilver', 'WinterBronze']].multiply([3, 2, 1]).sum(axis=1)
summer_score = medals[['SummerGold', 'SummerSilver', 'SummerBronze']].multiply([3, 2, 1]).sum(axis=1)
summer_per_game = summer_score / medals['SummerGame'].replace(0, 1)
winter_per_game = winter_score / medals['WinterGame'].replace(0, 1)
medals['PerGame'] = summer_per_game + winter_per_game
medals.sort_values(by='PerGame').head(10)
```
