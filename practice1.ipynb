```python
import pandas as pd
```

**1. โหลดไฟล์ชื่อ olympic_cleaned.csv ขึ้นมาใน DataFrame ชื่อ medals**

```python
medals = pd.read_csv('olympics_cleaned.csv')
```

**2. แสดงจนวนแถวและจนวนคอลมนของ medals**

```python
medals # shape = 146, 10
```

**3. ตรวจสอบวารหัสประเทศ (Code) ไมมีซํ้ากันดวย attribute is_unique**

```python
medals['Code'].nunique() == medals.shape[0]
```

**4. ตั้งรหัสประเทศใหเป็น index**

```python
medals.set_index('Code')
```

**5. แสดงคอลัมน์เหรียญทองโอลิมปิคฤดูร้อน (SummerGold)**

```python
medals[['Code', 'SummerGold']]
```

**6. แสดงข้อมูลของประเทศสหรัฐอเมริกา (USA)**

```python
medals[medals['Code'] == 'USA']
```

**7. แสดงเฉพาะข้อมลของโอลิมปิคฤดูร้อนของทุกประเทศ (คอลัมน์ที่ขึ้นต้นด้วยคำว่า Summer)**

```python
medals.iloc[:, 1:6]
```

**8. แสดงข้อมูลเหรียญทองของโอลิมปิกฤดูร้อนและฤดูหนาวของประเทศไทย (THA) สิงคโปร์ (SIN) และมาเล
เซีย (MAS) แสดงชื่อประเทศด้วย**

```python
medals.loc[medals['Code'].isin(['THA', 'SIN', 'MAS'])][['Country','SummerGold', 'WinterGold']]
```

**9. แสดงข้อมลเหรียญเงินและเหรียญทองแดงของโอลิมปคฤดูร้อนของประเทศฝรงเศส (FRA) และสเปน (ESP)
แสดงชื่อประเทศด้วย**

```python
medals.loc[medals['Code'].isin(['FRA', 'ESP'])][['Country', 'SummerSilver', 'WinterSilver', 'SummerSilver', 'WinterBronze']]
```

**10. แสดงประเทศที่เข้าร่วมโอลิมปิคฤดูร้อนมากกว่า 24 ครั้ง**

```python
medals.query('SummerGame > 24').iloc[:, :6]
```

**11. แสดงประเทศที่ได้เหรียญทองของโอลิมปิคฤดูร้อนมากกว่า 200 เหรียญ**

```python
medals.query('SummerGold > 200')
```
