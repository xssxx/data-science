# diff

```python
monthly_sale = df_house.resample('M').size()
monthly_sale
```

```python
monthly_sale.diff(1)
```

```python
monthly_sale.diff(3)
```

```python
monthly_sale.diff(-1)
```

# shift

```python
monthly_sale.shift(1)
```

```python
df_compare = pd.concat([monthly_sale, monthly_sale.shift(1)], axis=1)
df_compare.columns = ['monthly_sale', 'last_month_sale']
df_compare
```

```python
df_compare.monthly_sale / df_compare.last_month_sale
```
